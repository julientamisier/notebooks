```python
pv_prod= pd.DataFrame(columns=['Date','Total PV Production (kW)'], index = range(0,len(pv_prod_total)))
for i in range (0,len(pv_prod_total)):
    pv_prod['Date'][i] = pv_prod_total['Date'][i]
    pv_prod['Total PV Production (kW)'][i]=float(pv_prod_total['Total PV Production (kW)'][i])
```

```python
from datetime import timedelta
df= pd.DataFrame(columns=['Date','Hour','E_demand','E_PV','Price_furniture','Price_TURPE_E','Price_TURPE_P','Price_DT_mod','Export_price'], index = range(0,len(pv_prod)))
a=0
start_date = datetime.datetime(2017,12,31);
start_time = datetime.time(0,0);
j=0
for i in range (0,len(df)-1):
    if (i %24) == 0:
        start_date = start_date + timedelta(days=1)
        j=0
    df['Date'][i]=start_date
    start_time = datetime.time(j,0)
    j=j+1
    df['Hour'][i]= start_time
    if (pv_prod['Total PV Production (kW)'][i] < 0):
        df['E_PV'][i]=0
    else:
        df['E_PV'][i]=pv_prod['Total PV Production (kW)'][i]

    if pv_prod["Heure"][i] == ol_conso['Heure'][i-a]:
        df['E_demand'][i]=ol_conso['Total OL consumption (kW)'][i-a]
    else:
        df['E_demand'][i]=ol_conso['Total OL consumption (kW)'][i-a-1]
        a=a+1
        
```
